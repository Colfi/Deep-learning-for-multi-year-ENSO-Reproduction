```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

```python
sst = xr.open_dataset("../../DATA/HadISST_sst_2x2.nc")['sst']

sst
```

```python
lon = np.arange(0,360,5)
lat = np.arange(-55,61,5)

sst_inp = sst.interp(lon=lon,lat=lat)
```

```python
sst_inp[0].plot()
```

```python
sst_ds = xr.Dataset({"sst":sst_inp})
sst_ds.to_netcdf("../data/sst_inp.nc")
```
